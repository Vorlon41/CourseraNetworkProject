# T Tests

```python
import scipy.stats as stats
```

```python
a = st_data_2009_clean['Patent applications, residents']
b = st_data_2018_clean['Patent applications, residents']
stats.ttest_ind(a,b)
```

```python
# run the t test for multiple variables
for i in range(1,len(columns)):
    a = st_data_2009_clean[columns[i]]
    b = st_data_2018_clean[columns[i]]
    print(columns[i])
    print(stats.ttest_ind(a,b))
```

```python
#clean up the above results
# %s p-value: %s means print value as a string
for i in range(1,len(columns)):
    a = st_data_2009_clean[columns[i]]
    b = st_data_2018_clean[columns[i]]
    print(columns[i])
    statistic, pvalue = stats.ttest_ind(a,b)
    print("Statistic: %s p-value: %s" %(statistic, pvalue))
```

```python
#make the data easier to read with an if statement
for i in range(1,len(columns)):
    a = st_data_2009_clean[columns[i]]
    b = st_data_2018_clean[columns[i]]
    print(columns[i])
    statistic, pvalue = stats.ttest_ind(a,b)
    print("Statistic: %s p-value: %s" %(statistic, pvalue))
    if pvalue < 0.05:
        print("Significant")
    else:
        print("Non-Significant")
```
